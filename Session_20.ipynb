```python
import pandas as pd

# Loading the cleaned dataset
df = pd.read_csv("ford_car_dataset_clean.csv")

# Independent Features (X)
X = df.drop("price", axis=1)

# Dependent Feature (Y)
Y = df["price"]

# Printing the shape of X and Y
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
# Identifying categorical columns
cat_cols = X.select_dtypes(include="object").columns
print("Categorical Columns:")
print(cat_cols)

# Performing One-Hot Encoding
X_encoded = pd.get_dummies(X, columns=cat_cols)

# Converting encoded columns to integer type
X_encoded = X_encoded.astype(int)

# Displaying first 5 rows
print(X_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Numerical columns
num_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

# Create StandardScaler object
scaler = StandardScaler()

# Apply Standard Scaling
X_encoded[num_cols] = scaler.fit_transform(X_encoded[num_cols])

# Show first 5 rows
print(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    Y,
    test_size=0.33,
    random_state=42
)

# Print the shape of each dataset
print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Create the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Print the intercept
print("Intercept:")
print(model.intercept_)

# Print the coefficients
print("\nCoefficients:")
print(model.coef_)
```

```python
# Make predictions on the test data
y_pred = model.predict(X_test)

# Print first 10 predicted values
print("First 10 Predicted Values:")
print(y_pred[:10])

# Print first 10 actual values
print("\nFirst 10 Actual Values:")
print(y_test.iloc[:10].values)
```

```python
from sklearn.metrics import r2_score

# Calculate R2 Score
r2 = r2_score(y_test, y_pred)

# Print R2 Score
print("R2 Score:", r2)

# Interpretation: The model predicts car prices well and gives good results.
```

```python
import joblib

# Save model
joblib.dump(model, "LR_ford_car.pkl")

# Save scaler
joblib.dump(scaler, "scaler.pkl")

# Save columns
joblib.dump(X_encoded.columns, "columns.pkl")
```
